```python
import matplotlib.pyplot as plt
import numpy as np  # Import numpy for random data generation

plt.figure(figsize=(12, 8))
plt.suptitle("Graph")

# Graph 1
plt.subplot(3, 3, 1)
plt.title("Table1")
plt.plot([1, 3], [5, 7], "o", color="blue")
plt.xlabel("Blue color")
plt.ylabel("Blue color")

# Graph 2
plt.subplot(3, 3, 2)
plt.plot([1, 3, 5, 7, 9], [2, 6, 2, 6, 2], "*", mfc="orange", ls="-")
plt.plot([1, 3, 5, 7, 9], [6, 3, 5, 3, 6], ls="dashed")

# Graph 3
plt.subplot(3, 3, 3)
# Corrected: Added mec='black' for the marker edge color as requested.
plt.plot([1, 3, 5, 7], [3, 6, 3, 6], marker="o", mfc="yellow", mec='black', color="blue", ls="-")

# Graph 4
plt.subplot(3, 3, 4)
plt.title("Table3")
plt.plot([1, 3, 5, 7], [1, 3, 5, 7], ls="-")
plt.grid(True)


# Graph 5
# Generate data for 10 points
num_points = 10
# Random x and y coordinates to "spread out" the points
np.random.seed(42) # for reproducible results
x_coords = np.random.rand(num_points) * 10
y_coords = np.random.rand(num_points) * 10

# Create lists with 2 sizes and 2 colors, repeated for 10 points
point_sizes = [60, 180] * 5  # Two sizes, alternating
point_colors = ['skyblue', 'coral'] * 5 # Two colors, alternating

plt.subplot(3, 3, 5)
plt.scatter(
    x_coords,
    y_coords,
    s=point_sizes,
    c=point_colors,
    alpha=0.8,
    edgecolors='black'
)
plt.title("Scatter Plot")

# Graph 6
plt.subplot(3, 3, 6)
categories = ['Small', 'Medium', 'Large']
values = [ 1.5,0.7, 2.5]

plt.bar(categories, values, color="red")

# Graph 7
plt.subplot(3, 3, 7)
pie_data = [35, 20, 20, 25]
pie_labels = ['35', '20', '20', '25']
plt.pie(pie_data, labels=pie_labels, explode=[0.01]*4, startangle=90, shadow=True,
    colors=["lightgreen", "pink", "wheat", "cyan"])
plt.legend()

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
