```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 6])
plt.title('Sample Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot([1,2,3],[4,5,6])
plt.subplot(1,2,2)
plt.plot([4,5,6],[1,2,3])
plt.suptitle('Two Subplots')
```

```python
x=[1,4,2]
y=[7,8,9]
plt.plot(x,y,"o")
```

```python
plt.plot([3,4,2],[1,4,2],c="red",ls="dashed")
plt.plot(x,y,"o",c="blue",ls=":",ms=10,)
```

```python
plt.plot(x,y,"o",c="blue",ls=":",ms=10,)
plt.grid(y)
plt.plot(x,y,"o",c="blue",ls=":",ms=10,)
plt.grid(y)
```
